```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
databreach = pd.read_csv('database.csv')
vehicles = pd.read_csv('database2.csv')
```

```python
# Assuming "Year" is a categorical column in the databreach DataFrame
databreach_counts = databreach['Year'].value_counts().sort_index()

# Assuming "Model Year" is a categorical column in the vehicles DataFrame
vehicles_counts = vehicles['Model Year'].value_counts().sort_index()

# Print counts for data breaches
print("Data Breach Counts:")
print(databreach_counts)

# Print counts for vehicle occurrences
print("\nVehicle Counts:")
print(vehicles_counts)
```

```python
# Assuming "Year" is a categorical column in the databreach DataFrame
databreach_counts = databreach['Year'].value_counts().sort_index()

# Assuming "Model Year" is a categorical column in the vehicles DataFrame
vehicles_counts = vehicles['Model Year'].value_counts().sort_index()

# Create a figure and axis
fig, ax1 = plt.subplots()

# Plot Data Breach count on the bottom axis
ax1.plot(databreach_counts.index, databreach_counts, marker='o', linestyle='-', label='Data Breach', color='b')

# Set the y-axis label and color
ax1.set_xlabel('Year')
ax1.set_ylabel('Data Breaches in the State of Washington by Year', color='b')
ax1.tick_params('y', colors='b')

# Set x-axis limits for the bottom axis
ax1.set_xlim(2015, 2021)

# Create a second y-axis for the second line
ax2 = ax1.twinx()

# Plot Vehicle count on the second y-axis
ax2.plot(vehicles_counts.index - 2, vehicles_counts, marker='o', linestyle='-', label='Vehicles', color='r')

# Set the y-axis label and color for the second axis
ax2.set_ylabel('EV Count in the State of Washington by Year', color='r')
ax2.tick_params('y', colors='r')

# Show legend
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you already have your data loaded into databreach and vehicles dataframes

# Count occurrences for each unique year in databreach (extended until 2023)
databreach_counts = databreach['Year'].value_counts().sort_index()
databreach_counts = databreach_counts.reindex(range(2015, 2024), fill_value=0)

# Count occurrences for each unique year in vehicles
vehicles_counts = vehicles['Model Year'].value_counts().sort_index()
vehicles_counts = vehicles_counts.reindex(range(2017, 2024), fill_value=0)

# Create a new figure with dual y-axes
fig, ax1 = plt.subplots()

# Plot the databreach data on the left y-axis and bottom x-axis
ax1.plot(databreach_counts.index, databreach_counts.values, color='blue', marker='o', label='Databreach Count')
ax1.set_xlabel('Year')
ax1.set_ylabel('Databreach Count', color='blue')
ax1.tick_params('y', colors='blue')

# Create a second y-axis on the right for the vehicles data
ax2 = ax1.twinx()
ax2.plot(vehicles_counts.index, vehicles_counts.values, color='red', marker='o', label='Vehicle Count')
ax2.set_ylabel('Vehicle Count', color='red')
ax2.tick_params('y', colors='red')

# Set x-axis ticks for both axes and show the values along the bottom axis
ax1.set_xticks(range(2015, 2024))
ax1.set_xticklabels(range(2015, 2024))
ax2.set_xticks(range(2017, 2024))
ax2.set_xticklabels(range(2017, 2024))

# Add year labels to the top x-axis
ax2.xaxis.tick_top()
ax2.xaxis.set_label_coords(0.5, 1.05)
ax2.set_xticks(range(2017, 2024), minor=True)
ax2.set_xticklabels(range(2017, 2024), minor=True)

# Show the plot
plt.title('Databreach and Vehicle Counts Over Years')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you already have your data loaded into databreach and vehicles dataframes

# Count occurrences for each unique year in databreach (extended until 2023)
databreach_counts = databreach['Year'].value_counts().sort_index()
databreach_counts = databreach_counts.reindex(range(2015, 2024), fill_value=0)

# Count occurrences for each unique year in vehicles
vehicles_counts = vehicles['Model Year'].value_counts().sort_index()
vehicles_counts = vehicles_counts.reindex(range(2017, 2024), fill_value=0)

# Create a new figure with dual y-axes
fig, ax1 = plt.subplots()

# Plot the databreach data on the left y-axis and bottom x-axis as bars
ax1.bar(databreach_counts.index - 0.2, databreach_counts.values, width=0.4, color='#3498db', label='Databreach Count')
ax1.set_xlabel('Year')
ax1.set_ylabel('Databreach Count', color='#3498db')
ax1.tick_params('y', colors='#3498db')

# Plot the vehicles data on the right y-axis as bars
ax2 = ax1.twinx()
ax2.bar(vehicles_counts.index + 0.2, vehicles_counts.values, width=0.4, color='lightblue', label='Vehicle Count')
ax2.set_ylabel('Vehicle Count', color='lightblue')
ax2.tick_params('y', colors='lightblue')

# Set x-axis ticks for both axes and show the values along the bottom axis
ax1.set_xticks(range(2015, 2024))
ax1.set_xticklabels(range(2015, 2024))

# Add year labels to the top x-axis
ax2.xaxis.tick_top()
ax2.xaxis.set_label_coords(0.5, 1.05)
ax2.set_xticks(range(2017, 2024), minor=True)
ax2.set_xticklabels(range(2017, 2024), minor=True)

# Show the plot
plt.title('Databreach and Vehicle Counts Over Years')
plt.legend(loc='upper left', bbox_to_anchor=(0.7, 1))
plt.show()
```
