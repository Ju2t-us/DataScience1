```python
import requests
import pandas as pd
import seaborn as sns
import matplotlib as plt
```

```python
api_key = '31df2c87de2a41dd8a52c2f0fe934b4b'
base_url = 'https://api.themoviedb.org/3/'

actor_id = 5292

search_url = f'{base_url}person/{actor_id}/movie_credits?api_key={api_key}&language=en-US'
response = requests.get(search_url)
data = response.json()

movies = data.get('cast', [])

print(f"Arnold Schwarzenegger has acted in {len(movies)} movies.")
```

```python
movies_df = pd.DataFrame(movies)

plt.figure(figsize=(20, 6))
movies_df['rounded_popularity'] = round(movies_df['popularity'])

sns.countplot(x='rounded_popularity', data=movies_df)
```
