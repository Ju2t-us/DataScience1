```python
import requests
from bs4 import BeautifulSoup
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Generating a list of 100 random nouns
random_nouns = ["apple", "banana", "tree", "book", "car", "computer", "city", "ocean", "mountain", "music",
                "dog", "cat", "flower", "coffee", "building", "keyboard", "river", "island", "film", "guitar",
                "friend", "sky", "phone", "robot", "candle", "mirror", "clock", "shoe", "chair", "cloud",
                "house", "fire", "star", "painting", "camera", "hat", "pen", "window", "door", "key", "map",
                "globe", "mask", "mirror", "light", "train", "bicycle", "fruit", "vegetable", "musician", "artist",
                "teacher", "student", "scientist", "actor", "writer", "chef", "doctor", "engineer", "athlete",
                "piano", "garden", "beach", "forest", "lake", "moon", "sun", "planet", "space", "rocket",
                "microphone", "telescope", "cup", "plate", "spoon", "fork", "knife", "umbrella", "bag", "shovel",
                "treehouse", "fountain", "statue", "paintbrush", "palette", "glasses", "towel", "cloud", "moonlight",
                "sunrise", "sunset"]

# List to store the count of images for each article
image_counts = []

# Function to count images in a Wikipedia article
def count_images(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    # Counting images by finding all 'img' tags
    images = soup.find_all('img')
    return len(images)

# Iterate through each random noun
for noun in random_nouns:
    # Constructing the Wikipedia URL
    url = f"https://en.wikipedia.org/wiki/{noun}"
    
    try:
        # Counting images and appending to the list
        image_count = count_images(url)
        image_counts.append(image_count)
        print(f"Number of images in {noun}: {image_count}")
    except Exception as e:
        print(f"Error accessing {noun}: {e}")
        continue

# Calculating the average
average_image_count = sum(image_counts) / len(image_counts)

print(f"\nAverage number of images: {average_image_count}")
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.barplot(x=random_nouns, y=image_counts, palette="viridis")
plt.title("Number of Images in Wikipedia Articles")
plt.xlabel("Article")
plt.ylabel("Number of Images")
plt.show()
```
